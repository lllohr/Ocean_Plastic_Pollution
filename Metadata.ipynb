```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import requests
import io
import os
```

```python
# 1. Create a DataFrame for mlw_data 
mlw_data_df = pd.read_csv("Resources/MLW_Data.csv", encoding ="unicode_escape")
mlw_data_df.head(20)
```

```python
# 2. Check the datatypes of your columns. 
mlw_data_df.dtypes
```

```python
print(f'dataset has {mlw_data_df.shape[0]} rows and {mlw_data_df.shape[1]} columns.')
```

```python
#Get the columns and rows that are not null
mlw_data_df.count()
```

```python
#Determine if there are any missing values
mlw_data_df.isnull()
```

```python
mlw_data_df.drop(["CommunityName","BeachLocation", "EventDate","EventType","NatRef"], axis=1, inplace=True)
mlw_data_df.head()
```

```python
mlw_data_df.drop(["lon_x1", "lat_y1","lon_x2","lat_y2"], axis=1, inplace=True )
mlw_data_df.head(20)
```

```python
url="https://gist.githubusercontent.com/radcliff/f09c0f88344a7fcef373/raw/2753c482ad091c54b1822288ad2e4811c021d8ec/wikipedia-iso-country-codes.csv"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
c.head(3)
```

```python
# obtain country names from codes
c['CountryName'] = c[c.columns[0]]
c['CountryCode'] = c[c.columns[1]]
c = c[['CountryCode', 'CountryName']]
dictio = c.set_index('CountryCode').to_dict()['CountryName']
del dictio[np.nan]
mlw_data_df['BeachCountry']= mlw_data_df.BeachCountrycode.map(dictio)
mlw_data_df.drop('BeachCountrycode', axis=1, inplace=True)
mlw_data_df.head()
```

```python
mlw_data_df.drop_duplicates(inplace=True, ignore_index=True)
mlw_data_df.head()
```

```python
# shift column 'Name' to first position
first_column = mlw_data_df.pop("BeachCountry")
  
# insert column using insert(position,column_name,
# first_column) function
mlw_data_df.insert(0, "BeachCountry", first_column)

mlw_data_df.head()
  
```

```python
non_g_cols = [col for col in mlw_data_df.columns if 'G' not in col] 
g_cols = list(set(mlw_data_df.columns)-set(non_g_cols))
[(col, mlw_data_df[col].dtype) for col in non_g_cols]
mlw_data_df[non_g_cols].head(10)
```
