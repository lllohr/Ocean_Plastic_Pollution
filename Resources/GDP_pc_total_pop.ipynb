```python
import pandas as pd
```

```python
# 1. Create a DataFrame for plastic waste mismangement per capita vs GDP. 
PC_VS_GDP_df = pd.read_csv("Resources/Raw_Data/per-capita-plastic-waste-vs-gdp-per-capita.csv")
PC_VS_GDP_df.head()
```

```python
# 2. Check the datatypes of your columns. 
PC_VS_GDP_df.dtypes
```

```python
#Get the columns and rows that are not null
PC_VS_GDP_df.count()
```

```python
#Rename Columns
PC_VS_GDP_df.rename(columns={"Entity":"Country", "Per capita plastic waste (kg/person/day)": "Per_capita_plastic_waste", "GDP per capita, PPP (constant 2011 international $)": "GDP_per_capita","Total population (Gapminder, HYDE & UN)": "Total_population"}, inplace =True)
PC_VS_GDP_df.head()
```

```python
# Drop 
PC_VS_GDP_df.drop(["Per_capita_plastic_waste","Continent"], axis=1, inplace=True)
PC_VS_GDP_df.head()
```

```python
# Set null values to zero for GDP & Total pop
PC_VS_GDP_df["GDP_per_capita"] = PC_VS_GDP_df["GDP_per_capita"].fillna(0)
PC_VS_GDP_df["Total_population"] = PC_VS_GDP_df["Total_population"].fillna(0)
PC_VS_GDP_df.head(20)
```

```python
#Format columns 
PC_VS_GDP_df["GDP_per_capita"]= PC_VS_GDP_df["GDP_per_capita"].map("{:,.2f}".format)
PC_VS_GDP_df["Total_population"]= PC_VS_GDP_df["Total_population"].map("{:,.2f}".format)
PC_VS_GDP_df.head()
```

```python
print(f'dataset has {PC_VS_GDP_df.shape[0]} rows and {PC_VS_GDP_df.shape[1]} columns.')
```

```python
# Export dataframe as a new CSV file
PC_VS_GDP_df.to_csv("Resources/Clean_Data/GDP_pc_total_pop_clean.csv", index=False)
```
